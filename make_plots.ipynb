```python
from plot import load_outputs, plot_final_loss_vs_lr
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "serif"
plt.rcParams['font.size'] = 12
plt.rcParams['axes.linewidth'] = 1.5
plt.rc('text', usetex=False)
plt.rc('mathtext', fontset='cm')   # optional: use Computer Modern fontsplt.rc('legend', fontsize=10)
```

```python
output_dir = "gptopt/outputs/test_dap"
outputs = load_outputs(output_dir)

# 2) extract unique method names (before the "-lr-" suffix)
methods = sorted({ output['name'].split('-lr-')[0] for output in outputs })

# 3) get a discrete viridis with len(methods) colors
base_cmap = plt.get_cmap('viridis', len(methods))

# 4) map each method name â†’ RGBA tuple
colormap = { m: base_cmap(i) for i, m in enumerate(methods) }

# 5) plot
plot_final_loss_vs_lr(outputs, colormap, outfilename="test_dap", val=True)
plt.show()
```
